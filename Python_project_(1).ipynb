Q10. Please create a new column as "total_bedroom_size". If total_bedroom is 10 or less it should be named as SMALL. If the total_bedroom is 11 or more but less than 1000 it shouldd be named as MEDIUM, otherwise it should be considered LARGE.

```python
# for this we first define a new function and give all the condition

def tbs(x):
  bed_size= x[0]
  if (bed_size <= 10):    # condition for small bedroom
    return "SMALL"
  elif (bed_size >= 11) and (bed_size < 1000):   # condition for medium bedroom
    return "MEDIUM"
  else:
    return ("LARGE")    # for large bedroom
```

```python
# now we assign the above function to a new column b using .apply function also with the help of lambda.

housing["total_bedroom_size"] = housing[["total_bedrooms"]].apply(tbs,axis = 1)

housing
```

At the end of the table we can see that a new column has been created with name "total_bedroom_size".
